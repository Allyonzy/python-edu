# SQLITE3

```python
import sqlite3 as sl
import pandas as pd
```

```python
db_name = "test_connection.db" # указываем имя бд
con = sl.connect(db_name) # подключение
```

```python
data_to_db = pd.read_csv("forbes_richest_atheletes.csv", index_col="S.NO")
data_to_db
```

```python
data_to_db.to_sql('athletes', con=con)
```

```python
READ_ALL = "SELECT * FROM athletes;"
data = con.execute(READ_ALL)

for row in data:
  print(row)
```
